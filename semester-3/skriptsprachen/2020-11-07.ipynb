# Bubble Sort

1. Erzeuge eine List von Ganzzahlen von 0 bis 9 in aufsteigender Reihenfilge
1. Gib die list zur Kontrolle aus
1. Wende mit steigendem i, beim ersten Element beginnendm auf die Liste folgendes Verfahren an
   - TDB
1. TDB

```python
list_length = 10
# 1 Liste erzeugen
to_sort = tuple(range(list_length+1))

# 2 Ausgeben
print(to_sort)

def bubble_sort(l, sort_function=lambda x, y: x < y):
    sorted_list = list(l)
    is_sorted = False
    while not is_sorted:
        is_sorted = True
        for i in range(len(sorted_list)-1):
            if sort_function(sorted_list[i], sorted_list[i+1]):
                sorted_list[i], sorted_list[i+1] = sorted_list[i+1], sorted_list[i]
                is_sorted = False
    return sorted_list
                
print(tuple(bubble_sort(to_sort)))
```

# Komprimierung 2

- Dictionary so erstellen, dass man ein en und decoding machen kann

```python
from itertools import chain

input_file = open("moby.txt", "rt")

lines = tuple(input_file)
input_file.close()

words_unflattened = list(map(lambda line: line.split(), lines))
words = tuple(chain(*words_unflattened))

word_hist = dict()
for word in words:
    word_hist[word] = word_hist.get(word, 0) + 1
word_prios = sorted(word_hist.keys(), key=lambda word: word_hist[word], reverse=True)
word_codes = dict()
word_decodes = dict()
for code, word in enumerate(word_prios):
    word_codes[word]   = code
    word_decodes[code] = word

print(word_codes)
print(word_decodes)
```
